```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81            #acceleration of gravity in m/s^2
L= 0.3                #legth of the string in m

phi0= 5   #initial angle in degrees
phi= (phi0*np.pi)/180      #convering to radians 
omga = 0            #initial anglar velocity in rad/s
alpha = 0           #initial acceleration in rad/s^2

t=0                  #initial time in sec 
dt=0.0005              # integrating time step 

T = []
W = []
W2 =[]

phi0_2 = 85                            #initial angle in degrees
phi_2= (phi0_2*np.pi)/180              #convering to radians 
omga_2 = 0                             #initial anglar velocity in rad/s
alpha_2 = 0                            #initial acceleration in rad/s^2

while t<= 3:     #How long it will run 
    
    
    alpha = (-g/L)*np.sin(phi)
    omga = omga + alpha*dt
    phi = phi + (omga*dt) + (1/2)*((dt)**2)*alpha
    
    alpha_2 = (-g/L)*np.sin(phi_2)
    omga_2 = omga_2 + alpha_2*dt
    phi_2 = phi_2 + (omga_2*dt) + (1/2)*((dt)**2)*alpha_2
        
    
    W2.append(omga_2)
    T.append(t)
    W.append(omga)

    t = t + dt
     
phiplot = plt.plot(T,W,color='red',lw=1)
plt.title('Angular velocity vs Time (phi_0 = 5 deg)')
plt.xlabel('time (sec)')
plt.ylabel('angular velocity (rad/sec)')
plt.show(phiplot)

plot2 = plt.plot(T,W2,color='black',lw=1)
plt.title('Angular velocity vs Time (phi_0 = 85 deg)')
plt.xlabel('time (sec)')
plt.ylabel('angular velocity (rad/sec)')
plt.show(plot2)


```

```python
from pylab import rcParams
P=[]                       #Periods list 
theta_i_list =[]
d_theta= 0.0005  



theta_0 = 1 #Initial angle in degrees

theta_i = (theta_0*np.pi)/180

while theta_i <= np.pi/2:
    T3 =[]   #Time list 
    W3 =[]   #angV list
    t3 =0
    
    angV=0
    alp=0
    theta = theta_i

    while angV <= 0:     #How long it will run    
    
        alp = (-g/L)*np.sin(theta)
        angV = angV + alp*dt
        theta = theta + (angV*dt) + (1/2)*((dt)**2)*alp
        
        t3 = t3 + dt
        T3.append(t3)
        W3.append(angV)
            
    P.append(2*t3)
    theta_i_list.append(theta_i)
    theta_i = theta_i + d_theta
    

ang_array=np.asarray(theta_i_list) 
array2=ang_array*(180/np.pi)
new_theta_list = array2.tolist()
Ps = (2*np.pi)*(L/g)**(1/2)

Ps_array=np.ones(3107)*Ps
new_Ps_list=Ps_array.tolist()
 
plot3 = plt.plot(new_theta_list,P,color='blue',lw=2,label="Real T")
plot4 = plt.plot(new_theta_list,new_Ps_list,color='red',lw=2,label="small angle T")
plt.title('Period vs Initial angle')
plt.xlabel('Initial angle (deg)')
plt.ylabel('period (sec)')
rcParams['figure.figsize'] = 7, 7
plt.rcParams.update({'font.size': 16})
plt.legend(loc='upper left')
plt.show(plot3,plot4)


print('The period of oscillation for small angles is',round(Ps,3),'seconds.The length of the string is 0.3 meters.')
```
