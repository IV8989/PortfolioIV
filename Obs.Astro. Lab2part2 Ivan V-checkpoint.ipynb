```python
from astropy.io import fits
import astropy.units as u
from astropy.nddata import CCDData
from astropy.stats import sigma_clipped_stats, SigmaClip
from astropy.visualization import ImageNormalize, LogStretch
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator
import numpy as np
from photutils.background import Background2D, MeanBackground

# Show plots in the notebook
%matplotlib inline


B_filt = '/Users/User/Downloads/Bcomb.fits'
V_filt = '/Users/User/Downloads/Vcomb.fits'

with fits.open(B_filt) as Blist:
    Blist.info()
    data = Blist[0].data
    header = Blist[0].header
    
```

```python
# Define the mask. Removing all zero and negative counts. 
mask = data <= 0
```

```python
unit = u.electron / u.s
B_image = CCDData(data, unit=unit, meta=header, mask=mask)
```

```python
##Finding the min and max values for normalization
from astropy.visualization import MinMaxInterval
interval = MinMaxInterval()
interval.get_limits(B_image)
```

```python

# Set up the figure with subplots
fig, ax1 = plt.subplots(1, 1, figsize=(8, 8))

# Set up the normalization and colormap
norm_image = ImageNormalize(vmin=5e-8, vmax=8000, stretch=LogStretch(), clip=False)
cmap = plt.get_cmap('viridis')
cmap.set_over(cmap.colors[-1])
cmap.set_under(cmap.colors[0])
cmap.set_bad('white') # Show masked data as white
B_image_clipped = np.clip(B_image, 1e-8, 8000) # clip to plot with logarithmic stretch

# Plot the data
fitsplot = ax1.imshow(np.ma.masked_where(B_image.mask, B_image_clipped), 
                      norm=norm_image, cmap=cmap)

# Define the colorbar and fix the labels
cbar = plt.colorbar(fitsplot, fraction=0.07, pad=0.1)

# Define labels
cbar.set_label(r'Flux Count Rate ({})'.format(B_image.unit.to_string('latex')), 
                 rotation=270,labelpad=30)
ax1.set_xlabel('X (pixels)')
ax1.set_ylabel('Y (pixels)')
ax1.set_title('B filter image')
```

```python
from photutils import find_peaks
from photutils.centroids import centroid_2dg
```

```python
mean, median, std = sigma_clipped_stats(B_image.data, sigma=3.0, maxiters=5, mask=B_image.mask)
```

```python
from photutils import detect_threshold
threshold = detect_threshold(data, nsigma=20.)

sources_findpeaks = find_peaks(B_image.data, mask=B_image.mask, 
                               threshold=threshold, box_size=30, 
                               centroid_func=centroid_2dg)     
# Display the table
sources_findpeaks
```

```python
# Set up the figure with subplots
fig, ax1 = plt.subplots(1, 1, figsize=(8, 8))

# Plot the data
fitsplot = ax1.imshow(np.ma.masked_where(B_image.mask, B_image_clipped), norm=norm_image)
ax1.scatter(sources_findpeaks['x_centroid'], sources_findpeaks['y_centroid'], s=10, marker='.', 
            lw=1, alpha=0.7, color='r')#facecolor='None', edgecolor='r')

# Define the colorbar
cbar = plt.colorbar(fitsplot, fraction=0.07, pad=0.1)

# Define labels
cbar.set_label(r'Flux Count Rate ({})'.format(B_image.unit.to_string('latex')), 
                rotation=270,labelpad=30)
ax1.set_xlabel('X (pixels)')
ax1.set_ylabel('Y (pixels)')
ax1.set_title('find\_peaks Sources')
```

```python
from photutils import (detect_sources, source_properties, \
                       CircularAnnulus, CircularAperture)
# Define threshold and minimum object size
npixels = 10

# Create a segmentation image
segm = detect_sources(B_image.data, threshold, npixels)

# Create a catalog using source properties
catalog = source_properties(B_image.data, segm)
table = catalog.to_table()

# Display the table

table
```

```python
table['id','xcentroid','ycentroid','equivalent_radius']
```

```python
from photutils import CircularAperture
```

```python
# Create the apertures
circular_apertures = []
for obj in catalog:
    position = (obj.xcentroid.value, obj.ycentroid.value)
    radius = obj.equivalent_radius.value
    circular_apertures.append(CircularAperture(position, r=radius))
```

```python
# Set up the figure with subplots
fig, ax1 = plt.subplots(1, 1, figsize=(8, 8))

# Plot the data
fitsplot = ax1.imshow(np.ma.masked_where(B_image.mask, B_image_clipped), norm=norm_image)

# Plot the apertures
for aperture in circular_apertures:
    aperture.plot(color='red', alpha=0.7, axes=ax1)

# Define the colorbar
cbar = plt.colorbar(fitsplot, fraction=0.07, pad=0.1)


# Define labels
cbar.set_label(r'Flux Count Rate ({})'.format(B_image.unit.to_string('latex')), 
                 rotation=270,labelpad=30)
ax1.set_xlabel('X (pixels)')
ax1.set_ylabel('Y (pixels)')
ax1.set_title('Circular Apertures')
```

```python
##Performing Aperture Photometry 
from photutils import aperture_photometry
from astropy.table import QTable
```

```python
# The CCDData mask will be automatically applied
phot_datum = aperture_photometry(B_image, circular_apertures[0]) 
phot_datum
```

```python
# The CCDData mask will be automatically applied
phot_table = aperture_photometry(B_image, circular_apertures[0])
id = 1
for aperture in circular_apertures[1:]:
    id += 1
    phot_row = aperture_photometry(B_image, aperture)[0]
    phot_row[0] = id
    phot_table.add_row(phot_row)
```

## Photometry results for B image.

```python
# Display the table
phot_table
```

## Creating circular annuli

```python
# Create the apertures
circular_annuli = []
for obj in catalog:
    position = (obj.xcentroid.value, obj.ycentroid.value)
    r_in = obj.equivalent_radius.value + (0.25*obj.equivalent_radius.value)  
    r_out =obj.equivalent_radius.value + (0.5*obj.equivalent_radius.value)  

    circular_annuli.append(CircularAnnulus(position, r_in, r_out))
```

```python
# Set up the figure with subplots
fig, ax1 = plt.subplots(1, 1, figsize=(8, 8))

# Plot the data
fitsplot = ax1.imshow(np.ma.masked_where(B_image.mask, B_image_clipped), 
                      norm=norm_image)

# Plot the apertures
for aperture in circular_annuli:
    aperture.plot(color='yellow', alpha=0.4, axes=ax1, fill=True)
for aperture in circular_apertures:
    aperture.plot(color='red', alpha=0.7, axes=ax1)
# Define the colorbar
cbar = plt.colorbar(fitsplot, fraction=0.07, pad=0.1)


# Define labels
cbar.set_label(r'Flux Count Rate ({})'.format(B_image.unit.to_string('latex')), 
               rotation=270,labelpad=30)
ax1.set_xlabel('X (pixels)')
ax1.set_ylabel('Y (pixels)')
ax1.set_title('Circular Apertures and Annuli')
```

```python
##Calculating aperture corrections
# The CCDData mask will be automatically applied
bkg_phot_table = aperture_photometry(B_image, circular_annuli[0])
id = 1
for aperture in circular_annuli[1:]:
    id += 1
    phot_row = aperture_photometry(B_image, aperture)[0]
    phot_row[0] = id
    bkg_phot_table.add_row(phot_row)
```

## B Image annuli counts. 

```python
# Display table
bkg_phot_table
```

```python
# Calculate the mean background level (per pixel) in the annuli 
bkg_area = [annulus.area for annulus in circular_annuli]
bkg_mean_per_aperture = bkg_phot_table['aperture_sum'].value / bkg_area
bkg_mean = np.average(bkg_mean_per_aperture) * (u.electron / u.s)
print('Background mean per pixel:', bkg_mean)

# Calculate the total background within each circular aperture
bkg_sum = bkg_mean * table['area'].value

# Subtract the background from the original photometry
flux_bkgsub = phot_table['aperture_sum'] - bkg_sum

# Add this as a column to the original photometry table
phot_table['aperture_sum_bkgsub'] = flux_bkgsub
```

## Aperture Photometry, Background subtracted

```python
# Display table
phot_table
```

```python
##Converting to aparent magnitudes
t_dk=2.0254  #subintegration time sec

flux_bkgsub_nu= flux_bkgsub/(1*( u.electron / u.s))   #unitless flux
B_mag=-2.5*(np.log10(flux_bkgsub_nu))#+2.5*np.log10(t_dk)
phot_table['B magnitude']=B_mag
```
